# clean data for states

## remove columns, round numbers, and rename columns

```python
df = gpd.read_file('updated_states.geojson')
df = df.drop(columns=['Exp_Current_RESA', 'id', 'density'])
```

```python
#df = df.round(2)
df.columns = df.columns.str.replace('_', ' ')
df = df[df['name'] != 'Puerto Rico']
df = df.drop(16)
df
```

## Create poverty to expenditure metric
Revenue Per Student in Poverty: Higher values indicate more funding is available per student in poverty, which might suggest better resource allocation for high-need areas.

```python
df['PSI'] = (
    df['Total Revenue'] / df['Number of Students']
) / (df['Estimated School Age Poverty'] / df['Estimated School Age Population'])

df['Revenue per Student'] = df['Total Revenue'] / df['Number of Students']
df['Poverty Share'] = df['Estimated School Age Poverty'] / df['Estimated School Age Population']

#df['Poverty_Adjusted_Revenue'] = df['Poverty_Adjusted_Revenue'].apply(lambda x: x * 1e-6)
df.head()
```
